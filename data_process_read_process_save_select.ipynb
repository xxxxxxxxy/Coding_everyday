```python
import pandas as pd
```

```python
df_train = pd.read_csv('train.csv')
```

```python
df_train
```

```python
#统计某一列positive/negative/neutral各有多少
df_train.Overall_Sentiment.value_counts()
```

```python
#替换掉very的
#positive为1，neutral为0，negative为2
#替换的时候要先拷贝，否则会报错
df_copy=df_train.copy(deep=True)
i=0

while i < len(df_train):
    if df_copy['Overall_Sentiment'][i]=='very_positive' or df_copy['Overall_Sentiment'][i]=='positive':
        df_train['Overall_Sentiment'][i]=1
    elif df_copy['Overall_Sentiment'][i]=='very_negative' or df_copy['Overall_Sentiment'][i]=='negative':
        df_train['Overall_Sentiment'][i]=2
    elif df_copy['Overall_Sentiment'][i]=='neutral':
        df_train['Overall_Sentiment'][i]=0
    else:
        print(i)
    i=i+1
```

```python
df_train.Overall_Sentiment.value_counts()
```

```python
#把纠正过的文字加入
i=0
while i < len(df_train):
    if df_train['corrected_text'][i]==' ':
        df_train['corrected_text'][i] = df_train['OCR_extracted_text'][i]
    i=i+1
```

```python
#选出taskA需要的列
dict_df = {'label':df_train['Overall_Sentiment'], 'Text':df_train['corrected_text'], 'Image_name': df_train['Image_name']}
df_task_A = pd.DataFrame(dict_df)
```

```python
df_task_A.to_csv("taskA/Task_A_train.csv",index=False)
```
