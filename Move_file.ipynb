# 背景描述：

把一个文件夹的图片分到三个文件夹，分类依据是csv文件中的overall sentiment的标签

思路：

1.读取csv

2.将positive\negative\neutral对应的文件名称分别存储在三个list中：pos_list\neg_list\neu_list

3.使用shutil.move移动文件到对应的文件夹（使用到了os.path.join将地址进行连接）

Background:

Move file from one folder to another folder according to csv file information

Steps:

1.read csv

2.get positive\negative\neutral file name respectively according to csv file and save them to pos_list\neg_list\neu_list respectively

3.shutil.move

```python
import pandas as pd
df_train = pd.read_csv('data_7000.csv',header=None)
```

```python
#看看数据长什么样子
df_train.head()
```

```python
#将各个类别的图片名称读出来，放到不同的list中
i=0
pos_list=[]
neg_list=[]
neu_list=[]
while i<len(df_train):
    if df_train[8][i]=='positive':
        pos_list.append(df_train[0][i])
    elif df_train[8][i]=='very_positive':
        pos_list.append(df_train[0][i])
    elif df_train[8][i]=='negative':
        neg_list.append(df_train[0][i])
    elif df_train[8][i]=='very_negative':
        neg_list.append(df_train[0][i])
    elif df_train[8][i]=='neutral':
        neu_list.append(df_train[0][i])
    else:
        print(df_train[8][i])
    i=i+1
```

```python
#对train的图片分文件夹存放
import shutil,os
root_path = r'E:\研究生\伦敦玛丽女王大学短期访学\meme分类数据\data_7000'
pos_path = r'E:\研究生\伦敦玛丽女王大学短期访学\meme分类数据\data_7000\train_pos'#os.path.join(root_path,train_pos)
neg_path = r'E:\研究生\伦敦玛丽女王大学短期访学\meme分类数据\data_7000\train_neg'
neu_path = r'E:\研究生\伦敦玛丽女王大学短期访学\meme分类数据\data_7000\train_neu'
for i in df_train[0]:
    if i in pos_list:
        src_path = os.path.join(root_path,i)
        shutil.move(src_path,pos_path)
    elif i in neg_list:
        src_path = os.path.join(root_path,i)
        shutil.move(src_path,neg_path)
    elif i in neu_list:
        src_path = os.path.join(root_path,i)
        shutil.move(src_path,neu_path)
```
